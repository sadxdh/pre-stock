```python
import efinance as ef
import time
from datetime import datetime, timedelta
import pandas as pd

# 股票代码
stock_code = '600519'
# 数据间隔时间为1分钟
freq = 1

# 初始化数据存储列表
data_list = []

# 获取当前时间
current_time = datetime.now()
# 确保在当前分钟的第一秒开始获取数据
current_time = current_time.replace(second=0, microsecond=0)

# 设置交易所闭市时间
sse_close_time_start = current_time.replace(hour=15, minute=0, second=0)  # 上海证券交易所闭市时间
sse_close_time_end = current_time.replace(hour=15, minute=30, second=0)  # 上海证券交易所闭市时间

szse_close_time_start = current_time.replace(hour=15, minute=0, second=0)  # 深圳证券交易所闭市时间
szse_close_time_end = current_time.replace(hour=15, minute=30, second=0)  # 深圳证券交易所闭市时间

hkex_close_time_start = current_time.replace(hour=16, minute=0, second=0)  # 香港证券交易所闭市时间
hkex_close_time_end = current_time.replace(hour=16, minute=30, second=0)  # 香港证券交易所闭市时间

while True:
    try:
        current_time = datetime.now()
        
        if sse_close_time_start <= current_time <= sse_close_time_end or \
           szse_close_time_start <= current_time <= szse_close_time_end or \
           hkex_close_time_start <= current_time <= hkex_close_time_end:
            print("交易所已闭市，停止运行脚本。")
            break

        # 获取当前分钟的数据
        df = ef.stock.get_quote_history(stock_code, klt=freq, start=current_time)
        
        if df.empty or len(df) <= 1:
            print(f'{stock_code} 已收盘')
            break
        else:
            df['涨跌幅'] = (df['收盘'] - df['开盘']) / df['开盘'] * 100
            df['涨跌额'] = df['收盘'] - df['开盘']
            data_list.append(df.iloc[-1])  # 选择最后一行数据（最新数据）

            print(df.iloc[-1])  # 输出最新的一行数据
            time.sleep(60)

            # 将数据保存到文件，此处以每30分钟保存一次为例
            if current_time.minute % 30 == 0:
                final_data = pd.DataFrame(data_list)
                file_name = f'{stock_code}_modified.csv'
                final_data.to_csv(file_name, encoding='utf-8-sig', index=None)
                print(f'数据已保存到文件: {file_name}')

    except Exception as e:
        print("发生异常：", e)
        time.sleep(60)

# 循环结束时保存剩余数据
if data_list:
    final_data = pd.DataFrame(data_list)
    file_name = f'{stock_code}_modified.csv'
    final_data.to_csv(file_name, encoding='utf-8-sig', index=None)
    print(f'最终数据已保存到文件: {file_name}')
```
